Блок 1. Индивидуальное задание 1

```python
def work(n):
  if n <= 5 or n >= 100:
    return 0

  list_1 = []
  for i in range(3, n):
    for j in range(2, i):
      if i % j == 0:
        break
    else:
      list_1.append(i)

  list_2 = []
  for i in range(200, 501):
    for j in range(2, i):
      if i % j == 0:
        break
    else:
      list_2.append(i)

  list_1 += list_2
  del list_1[20]

  return list_1[14] + list_1[30]

print(work(7))
print(work(17))
```

Блок 1. Индивидуальное задание 2

```python
def work(n):
  if n % 2 != 0:
   return 0

  set_1 = set()
  for i in range(2, n + 1, 2):
    set_1.add(i)

  set_2 = set()
  for i in range(7, 99, 7):
    set_2.add(i)

  sum_1 = sum(set_1.union(set_2))
  sum_2 = sum(set_1.intersection(set_2))
  sum_3 = sum(set_1.difference(set_2))

  return sum_1 + sum_2 + sum_3

print(work(20))
print(work(40))
```

Блок 1. Индивидуальное задание 3

```python
def shop(fruit_1, fruit_2):
  fruit_shop_1 = { 'банан': 15, 'кокос': 20, 'ананас': 40, 'апельсин': 30, 'мандарин': 25, 'яблоко': 20, 'киви': 25, 'арбуз': 40, 'дыня': 30 }
  fruit_shop_2 = { 'банан': 15, 'кокос': 25, 'ананас': 35, 'апельсин': 35, 'мандарин': 30, 'яблоко': 10, 'киви': 35, 'арбуз': 30, 'дыня': 40 }

  fruit_shop_1_min_key = 'банан'
  for key in fruit_shop_1:
    if (fruit_shop_1[key] < fruit_shop_1[fruit_shop_1_min_key]):
      fruit_shop_1_min_key = key
    fruit_shop_1[key] += 5

  del fruit_shop_1[fruit_shop_1_min_key]
  fruit_shop_1.update({ 'груша': 25 })

  fruit_shop_2_max_key = 'банан'
  for key in fruit_shop_2:
    if (fruit_shop_2[key] > fruit_shop_2[fruit_shop_2_max_key]):
      fruit_shop_2_max_key = key
    fruit_shop_2[key] -= 5

  del fruit_shop_2[fruit_shop_2_max_key]
  fruit_shop_2.update({ 'груша': 8 })

  return min(fruit_shop_1[fruit_1] + fruit_shop_1[fruit_2], fruit_shop_2[fruit_1] + fruit_shop_2[fruit_2])

print(shop('апельсин', 'мандарин'))
print(shop('апельсин', 'киви'))
```

Блок 1. Индивидуальное задание 4

```python
import sys

def check_generator():
  mas_gen = [i * 2 for i in range(1, 50000)]
  mem_1 = sys.getsizeof(mas_gen)

  def generator_function():
    for i in range(1, 50000):
        yield i * 2

  gen = generator_function()
  next(gen)
  mem_2 = sys.getsizeof(next(gen))

  return mem_1 - mem_2

print(check_generator())
```

Блок 1. Индивидуальное задание 5

```python
from functools import reduce

def function(n):
  if n <= 2:
    return 0

  orig_list = [i for i in range(10, 10 ** n + 10, 10)]

  list_1 = list(filter(lambda x: x % 8 == 0, orig_list))
  list_2 = list(map(lambda x: x ** 0.5, list_1))
  list_3 = list(zip(list_1, list_2))

  return round(reduce(lambda x, y: x + sum(y), list_3, 0))

print(function(5))
print(function(6))
```
