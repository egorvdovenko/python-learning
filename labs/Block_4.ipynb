```python
with open(gc_folder + '/lab_works_files/Карелия_Википедия.html') as f:
  soup = BeautifulSoup(f)

theaters_regex = re.compile(r'театр', re.IGNORECASE)
theaters_anchors = soup.find_all('a', { 'title': theaters_regex })

theaters_dict = {}
for ta in theaters_anchors:
    theater_name = ta.get('title')
    theater_link = ta.get("href")
    theaters_dict[theater_name] = theater_link

print(theaters_dict['Музыкальный театр Республики Карелия'])
```
